```python
# imports

import requests
import json
import pathlib
```

```python
# setup
base_url = "https://api.daikinskyport.com"
api_session = requests.session()
api_session.headers.update({
    "Accept": "application/json",
    "Content-Type": "application/json",
})
secrets = json.loads(pathlib.Path('secrets.json').read_text())
credentials = json.dumps(secrets['credentials'])
```

```python
# get access token
token_endpoint = base_url + '/users/auth/login'
api_res = api_session.post(token_endpoint, data=credentials).json()
token = api_res['accessToken']
refresh_token = api_res['refreshToken']
# add auth token to headers
api_session.headers.update({"Authorization": f'Bearer {token}'})
```

```python
# verify user
verify_user_endpoint = base_url + '/users/me'
r = api_session.get(verify_user_endpoint)
r.json()
```

```python
locations_endpoint = base_url + '/locations'
locations = api_session.get(locations_endpoint).json()
locations
```

```python
# get devices
devices_endpoint = base_url + '/devices'
devices = api_session.get(devices_endpoint).json()
devices
```

```python
device_data_endpoint = base_url + f"/deviceData/{devices[0]['id']}"
device_data = api_session.get(device_data_endpoint).json()
device_data
```
